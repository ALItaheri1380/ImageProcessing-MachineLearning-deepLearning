```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("D:/par/Deep learning/image.png")
gry_image = cv.imread("D:/par/Deep learning/image2.png" , 0)
```

```python
_ , tresh_pic = cv.threshold(gry_image , 230 , 255 , cv.THRESH_BINARY)
```

```python
plt.imshow(tresh_pic)
```

```python
final_image = cv.inpaint(image , tresh_pic , 10000 , flags=cv.INPAINT_TELEA)
```

```python
cv.imshow("image1" , image)
cv.imshow("image2" , final_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
